```python
import pandas as pd
```

```python
countries_df = pd.read_csv("UNSD_country_to_regions.csv")
```

```python
countries_df.head()
```

```python
countries_df.drop(columns=["Global Code", "Global Name", "Region Code", "Sub-region Code", "Intermediate Region Code",
                           "M49 Code", "ISO-alpha2 Code", "Least Developed Countries (LDC)", "Land Locked Developing Countries (LLDC)",
                           "Small Island Developing States (SIDS)", "Unnamed: 15"],axis=1,inplace=True)
countries_df.head()
```

```python
from sqlalchemy import create_engine
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column, Integer, String, Float
```

```python
Base = declarative_base()
```

```python
class Country(Base):
    __tablename__ = 'countries'
    id = Column(Integer, primary_key=True)
    country = Column(String(100))
    intermediate_region = Column(String(100))
    sub_region = Column(String(100))
    region = Column(String(100))
```

```python
#Create db connection
engine = create_engine("sqlite:///countries.sqlite")
conn = engine.connect()
```

```python
#creates tables from python classes
Base.metadata.create_all(engine)
```

```python
#create connection to edit db
from sqlalchemy.orm import Session
session = Session(bind=engine)
```

```python
#for loop this with scraped data to add to db?
for row in countries_df.itertuples():
    country_toadd = Country(country=row[4], intermediate_region=row[3], sub_region=row[2], region=row[1])
    session.add(country_toadd)
```

```python
#push additions to db
session.commit()
```

```python
#close session
session.close()
```

```python
session.query(Country.country).all()
```
