```python
from bs4 import BeautifulSoup as bs
import pandas as pd
import time
from webdriver_manager.chrome import ChromeDriverManager
from splinter import Browser
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Visit houseplantsexpert.com website
url = 'https://www.houseplantsexpert.com/a-z-list-of-house-plants.html'
browser.visit(url)
# time.sleep(5)
```

# Screen-scraping

```python

# common name, Scientific name, picture url (thumbnail), description, origin, names, temperature, picture URL (big)
plant_list=[]

for i in range(0,10):
    plant_dict = {}
    
    html = browser.html
    soup = bs(html, "html.parser")
    
    plant_data = soup.find_all('li', class_="simplePagerPage1")
    plant_dict["common_name"] = plant_data[i].find('a').text
    print( plant_dict["common_name"])
#     plant_dict["common_name"] = plant_data[i].a['data-name']
    
#     plant_dict["details_url"] = plant_data[i].a['href']
    plant_dict["image_url"] = plant_data[i].img['src']
        
    browser.links.find_by_partial_text(plant_data[i].find('a').text).click()
    
    time.sleep(1)
    
    html = browser.html
    soup = bs(html, "html.parser")
    
    plant_data2 = soup.find_all('div', class_="container")

    plant_dict["origin"] = plant_data2[3].find_all('td')[1].text
    plant_dict["scientific_name"] = plant_data2[3].find_all('td')[3].text.split('.')[1].split('(')[0]
    plant_dict["temperature"] = plant_data2[4].find_all('td')[1].text
    plant_dict["big_image_url"] = plant_data2[3].find_all('img')[0]["src"]
    
    
    plant_list.append(plant_dict)
    print(i)
    browser.back()
    time.sleep(1)
print(plant_list)
```

```python
html = browser.html
soup = bs(html, "html.parser")

# common name, Scientific name, picture url (thumbnail), description, origin, names, temperature, picture URL (big)

plant_data=[]

datas = soup.find('ul', class_="pagination1")
results = datas.find_all('a')

count = 0

for result in results:
    try:
        plant_dict = {}
        plant_dict["Common Name"] = result.text.lstrip()
        print(plant_dict["Common Name"])
        
    #     plant_dict["plant_url"] = datas.find_all('a')[count]['href']
        browser.visit(datas.find_all('a')[count]['href'])
        count += 1
        
        html = browser.html
        soup = bs(html, "html.parser")

        plant_data2 = soup.find_all('div', class_="container")

        plant_dict["origin"] = plant_data2[3].find_all('td')[1].text
        plant_dict["scientific_name"] = plant_data2[3].find_all('td')[3].text.split('.')[1].split('(')[0]
        plant_dict["temperature"] = plant_data2[4].find_all('td')[1].text
        plant_dict["big_image_url"] = plant_data2[3].find_all('img')[0]["src"]

        plant_data.append(plant_dict)
        browser.back()
       
    except:
        print("error")
        print(plant_dict["Common Name"])
        browser.back()
        
plant_data
```

```python
browser.quit()
```

```python
# store the data into dataframe
df = pd.DataFrame(plant_data)
df.head()
```

```python
df["temperature"].replace("°","", inplace = True)
df["temperature"]
```

```python
df.to_csv("plant_list.csv", index=True, header = True)
```
